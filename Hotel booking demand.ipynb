## Datetime values

```python
# datetime values are encountered constantly in data science projects
# in this dataset we have columns with year, month, and day, but the reservation_status_date column is an entire date
# from using info() earlier we can see this is currently being stored in the data frame as an object
# this means the dates are stored as string values
# we need to convert this to a datetime column, as date is easier to work with and manipulate rather than struggling with it as a string
# to do this, we use the to_datetime() function
data_df["reservation_status_date"] = pd.to_datetime(data_df["reservation_status_date"])
```

```python
# checking if it has been changed
data_df.info()
```

```python
# the [ns] here indicates that's a nanosecond-based time format, telling us the precision of the format
# we can do multiple types of analysis once we've converted our date into the correct format
# we can find the minimum and maximum date value, slice the values to look at particular instances between dates, etc.
```

```python
# get year, month, date from our date column
# to extract the year we'll create a new column and name it appropriately (reservation_status_year), then extract the value to it
data_df["reservation_status_year"] = data_df["reservation_status_date"].dt.year

#and do the same for month and day
data_df["reservation_status_month"] = data_df["reservation_status_date"].dt.month
data_df["reservation_status_day"] = data_df["reservation_status_date"].dt.day
data_df.head()
```

```python
# if we want, we can even extract the week of the year, or day of the week as well
# note: original tutorial used dt.week, but that is since deprecated, so dt.isocalendar().week was used instead as it's now correct
data_df["res_week_of_year"] = data_df["reservation_status_date"].dt.isocalendar().week
data_df["res_day_of_week"] = data_df["reservation_status_date"].dt.dayofweek
data_df.head()
```

```python
# once we've extracted everything we need from the datetime column, we can safely drop it
data_df.drop("reservation_status_date", inplace=True, axis=1) 
```

## Dealing with categorical data

```python
# creating a data frame with all the categorical features from our original data frame
# first, create a list of all the columns in the original data frame that is type object using list comprehension
cat_cols = [col for col in data_df.columns if data_df[col].dtype=="object"]
cat_cols
```

```python
# filter out a categorical feature data frame from our original data frame
cat_df = data_df[cat_cols]
cat_df.head()
```

```python
# printing unique values of each column
for col in cat_df.columns:
    print(f"{col}: \n{cat_df[col].unique()}\n")
```

```python
# encode arrival date month
# we will manually map each of these months to a number, starting with 1 for january
# using dictionary and map()
# map() takes a dictionary as an argument and returns a new column the same length as the original column,
# in this instance replacing month names with the corresponding numerical values
cat_df["arrival_date_month"] = cat_df["arrival_date_month"].map({"January":1, "February":2, "March":3, "April":4, "May":5, "June":6,
                                                                    "July":7, "August":8, "September":9, "October":10, "November":11, 
                                                                 "December":12})
```

```python
cat_df.head()
```

```python
# working on country and hotel columns
# country can't be one-hot encoded since it has so many unique values, will create a ton of new columns, so we'll label encode
# hotel has only 2 unique values, so we can use label encoding also, those will end up encoded as 0 and 1
# import LabelEncoder class from scikit learn (python data science library)
from sklearn.preprocessing import LabelEncoder

# initialize label encoder class
le = LabelEncoder()

# use fit_transform() from LabelEncoder on country feature
# fit_transform() does two things: 
# 'fit' component takes in the entire feature and maps each unique categorical value to a numerical value
# 'transform' makes the intended change
cat_df["country"] = le.fit_transform(cat_df["country"])

# same thing for hotel feature
cat_df["hotel"] = le.fit_transform(cat_df["hotel"])
```

```python
cat_df.head()
```

```python
# for the rest of the features we'll use one-hot encoding
# using get_dummies() from pandas to perform one-hot encoding
# we give the function the data frame as the input, and all the columns that need to undergo one-hot encoding
cat_df = pd.get_dummies(data=cat_df, columns=["meal", "market_segment", "distribution_channel",
                                             "reserved_room_type", "assigned_room_type", "customer_type", 
                                             "deposit_type", "reservation_status"])
```

```python
cat_df
```

```python
# creating a numerical data frame by dropping all the categorical features
# we could do this to our original data frame, but we're doing it this way to avoid any sort of confusion with the data frame means
num_df = data_df.drop(columns=cat_cols, axis=1)
```

```python
num_df
```

```python
# create final data frame by concatenating the two data frames
# axis indicates the axis we want to concatenate along, 0 for row-wise, 1 for column-wise
final_df = pd.concat([num_df, cat_df], axis=1)
```

## Correlation Analysis

```python
# create a correlation matrix using pandas corr()
# corr() calculates the correlation coefficients between all pairs of features in the data frame 
# and returns a data frame with the same shape as the input data frame
# absolute function ( abs() ) makes every value absolute, as correlation coefficients can be negative 
corr_matrix = final_df.corr().abs()
corr_matrix
```

```python
# setting the correlation coefficient threshold
# which anything over will be considered highly correlated 
# this value is open to modifications based on trial and error
threshold = 0.85
```

```python
# create a mask to identify highly correlated features

import numpy as np

# np.ones() creates an array of 1s with the same shape as corr_matrix
# np.triu() returns the upper triangular portion of this array, excluding the diagonal elements and everything below it
# here, k controls what elements should be zeroed out with respect to the diagonal
# k = 0 is the main diagonal; k=1 excludes the main diagonal and everything below it
# the resulting array is then cast to a Boolean array using astype(), so elements in the upper triangular portion are true,
# while elements in the lower triangular portion are false
# the resulting upper triangular array is then used as a mask to select the upper triangular portion of corr_matrix using where()
# where() sets all values in corr_matrix that correspond to false in the upper triangular to NaN
# the resulting matrix stored in the upper variable is a copy of corr_matrix with all elements in the lower triangular portion set to NaN
# we exclude diagonal elements in the lower portion of the triangle because diagonal values will always be 1 because that's 
# the correlation of the feature with itself
# everything below the diagonal is excluded because it's a mirror image of the upper triangular as correlation is a symmetric measure

upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
```

```python
# ensuring only 1 of the highly correlated features gets dropped
# first create an empty list called to_drop to keep track of the features we want to drop from the data frame
to_drop = []

# then, iterate over the upper triangle of the correlation matrix using two nested loops, one for rows and one for columns
# for each pair of features, check if the absolute value of the correlation coefficient is greater than or equal to the threshold
# if it is, we add the name wih the second feature of the pair, the one with the higher index, to the to_drop list
for i in range (len(corr_matrix.columns)):
    for j in range(i):
        if abs(corr_matrix.iloc[i, j]) >= threshold:
            colname = corr_matrix.columns[i]
            to_drop.append(colname)
```

```python
to_drop
```

```python
# drop all the columns in the list from the data frame
final_df.drop(columns=to_drop, inplace=True)
```

```python
# returning to agent feature from 'missing values' section
# looking at how many missing values it has
final_df.agent.isnull().sum()
```

```python
# in this case, we don't want to just delete the column, and filling with guesses would be difficult to justify
# it can be useful to look online for additional resources, or talk to your organization's domain expert
# with this dataset we have an article that the dataset is based on
# in this article they clarify for some categorical variables in this dataset, NULL should not be interpreted as 
# missing data but rather as 'not applicable'
# so for agent, a NULL value indicates the booking did not come from a travel agent

# in this case we just need to impute the missing values in the agent column with a 0 so a machine learning model
# or person looking at the data can identify it as a separate class
final_df["agent"] = final_df["agent"].fillna(0)
```
